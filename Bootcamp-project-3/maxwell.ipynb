```python
#Import the dependencies
import pandas as pd
from bs4 import BeautifulSoup as bs
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import requests
import geocoder
import numpy as np
```

```python
url = 'https://www.leadingrestaurants.co.uk/data/full-lists/hardens/'

# Retrieve pages with request
response = requests.get(url)
```

```python
# Create BS object; parse with html.parser
soup = bs(response.text, 'html.parser')
```

```python
# Print all paragraph texts
paragraphs = soup.find_all('div')

for paragraph in paragraphs:
    print(paragraph.text)
    print("----")
```

```python
def create_column(data, column_name: str, results):
    data[column_name] = results
    return data

def find_div(class_name, soup=soup):
    return soup.find_all('div', class_=class_name)

def create_list(results):
    return [res.text.strip("\n").replace("\n", "").strip(" ") for res in results]

data = pd.DataFrame()

class_names = ["res_name", "res_location", "top_pos"]
columns = ["restaurant_name", "location", "rank"]

for class_name, column in zip(class_names, columns):
    results = find_div(class_name)
    results = create_list(results)
    data = create_column(data, column, results)
    
data
```

```python
data.set_index("rank", inplace=True)
```

```python
data
```

```python
# get the links from the home page
results = soup.find_all('div', class_='res_name')

links = []

for res_name in results:
    link = res_name.findAll("a")
    
    if link:
        links.append(link[0]["href"])
    else:
        links.append("")
```

```python
#Printing the links to see which restaurants are closed
links
```

```python
#getting the address and rating 

import time 

address_items = ["streetAddress", "addressLocality", "addressRegion", "postalCode"]
address = []
ratings = []

for link in links:
    if link != "":
        html_page = requests.get(link)
        html_soup  = bs(html_page.text, "html.parser")
        
        rating = html_soup.findAll("span", style="color: #e7711b")
        if rating:
            ratings.append(float(rating[0].text))
        
        address_parts = []
        
        for address_item in address_items:
            item = html_soup.find("span", itemprop=address_item)
            if item:
                address_parts.append(item.text)
        
        res_add = ", ".join(address_parts)
        address.append(res_add)
    
    else:
        address.append("closed")
        ratings.append("closed")
    time.sleep(0.3)
```

```python
#Print the address
address
```

```python
#Print the ratings
ratings
```

```python
#Adding the address to the dataframe
data["address"] = address
```

```python
#Adding the ratings to the dataframe
data["ratings"] = ratings
```

```python
#get the latitude and longitude
def get_latlng(address):
    url = "https://nominatim.openstreetmap.org"
    g = geocoder.osm(address, url=url)
    try:
        res = g.json["lat"], g.json["lng"]
    except TypeError:
        res = np.nan
    return res
```

```python
data["lat_long"] = data.apply(lambda row: get_latlng(row["address"]), axis=1)
```

```python
data["lat"], data["long"] = data["lat_long"].str
```
